```python
#Importe de librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker
```

```python
#Creación de dataframe
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vS11cEHpxXSMKnAc7tpmImjYJvlF49frSnuJHTQBt-EAo_HlSzRm8ps4Wj0xtzNXMaxgBiQqppRFCOp/pub?gid=377071735&single=true&output=csv')
df.info()
```

```python
#Cmabio de tipo de variable "Localidad"
df['Código Localidad']=df['Código Localidad'].astype('Int64')
df
```

```python
#Cantidad de registros y columnas
df.shape
```

```python
#Descriptivos de las variables numericas de la base
df.describe()
```

```python
#Primeros 5 registros del dataset
df.head()
```

```python
#Tabla cruzada para comparar variable año y grupo condición crónica
df_causas=pd.crosstab(df['Año'],df['Grúpo condición cronica'])
df_causas
```

```python
#Grafica de pie para enfermedades cardiovasculares por año
fig,ax=plt.subplots(figsize=(15,4))
color=['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'gray', 'orange', 'purple', 'brown', 'pink']
ax.pie(df_causas['Enfermedades cardiocerebrovasculares'],autopct="%0.1f%%", colors=color)
labels=df_causas.index
ax.legend(labels, loc='center left', bbox_to_anchor=(1,0,0.5,1))
```

```python
#Cambio de categorias variable para visualización top 7
df.loc[df['Descripción causa básica de muerte'].str.contains('INFARTO AGUDO DEL MIOCARDIO, SIN OTRA ESPECIFICACION'), 'Descripción causa básica de muerte'] = 'Infarto miocardio'
df.loc[df['Descripción causa básica de muerte'].str.contains('HEMORRAGIA SUBARACNOIDEA, NO ESPECIFICADA'), 'Descripción causa básica de muerte'] = 'Hemorragias cerebrales'
df.loc[df['Descripción causa básica de muerte'].str.contains('HEMORRAGIA INTRAENCEFALICA, NO ESPECIFICADA'), 'Descripción causa básica de muerte'] = 'Hemorragias cerebrales'
df.loc[df['Descripción causa básica de muerte'].str.contains('ENFERMEDAD ATEROSCLEROTICA DEL CORAZON'), 'Descripción causa básica de muerte'] = 'Aterosclerotica del corazon'
df.loc[df['Descripción causa básica de muerte'].str.contains('ACCIDENTE VASCULAR ENCEFALICO AGUDO, NO ESPECIFICADO COMO HEMORRAGICO O ISQUEMICO'), 'Descripción causa básica de muerte'] = 'Accidente cerebro-vascular'
df.loc[df['Descripción causa básica de muerte'].str.contains('OTRAS ENFERMEDADES CEREBROVASCULARES ESPECIFICADAS '), 'Descripción causa básica de muerte'] = 'Otras enfermedades cardiovasculares'
df.loc[df['Descripción causa básica de muerte'].str.contains('INFARTO CEREBRAL, NO ESPECIFICADO'), 'Descripción causa básica de muerte'] = 'Infarto cerebral'
#Agrupamiento de otras categorias de la variable a "Otras enfermedades cardiovascuales"
top_enfermedades = ['Infarto miocardio', 'Hemorragias cerabrales', 'Aterosclerotica del corazon','Accidente cerebro-vascular','Enfermedad cardiaca','Otras enfermedades cardiovasculares','Infarto cerebral']
```

```python
#Funcion lambda para cambiar etiquetas de categoria
df['Descripción causa básica de muerte'] = df['Descripción causa básica de muerte'].apply(lambda x: 'Otras enfermedades cardiovasculares' if x not in top_enfermedades else x)
```

```python
#Tabla de frecuencias "Causa de muerte"
causas_muerte = df['Descripción causa básica de muerte'].value_counts()
causas_muerte = pd.DataFrame(causas_muerte)
causas_muerte.rename(columns={'count': 'Frecuencia'}, inplace=True)
causas_muerte['Frecuencia Rel.%'] = (causas_muerte['Frecuencia']/len(df))
causas_muerte
```

```python
#Tabla cruzada por régimen de aseguramiento y causa de muerte
df_causas=pd.crosstab(df['Régimen aseguramiento'],df['Descripción causa básica de muerte'])
df_causas
```

```python
#Grafico de barras distribucion por régimen de aseguramiento
fig, ax = plt.subplots(figsize = (14, 4))
ax = df['Régimen aseguramiento'].value_counts().plot(kind='bar', ax = ax)
plt.xlabel('Tipo de régimen de aseguramiento')
plt.ylabel('Frecuencia')
plt.xticks(rotation=0)
plt.xticks(fontsize=9)
plt.title("Caracterizacion de fallecidos por régimen de aseguramiento",fontsize=15)
#Bucle para que me muestre la etiqueta de cada columna
for container in ax.containers:
    ax.bar_label(container);
```

```python
#Tabla cruzada para régimen de aseguramiento por genero
df_regimen_sexo = pd.crosstab(df['Régimen aseguramiento'], df['Sexo'])
df_regimen_sexo
```

```python
#Grafico de barras distribucion de régimen de aseguramiento y genero
fig, ax = plt.subplots(figsize = (14, 4))
ax = df_regimen_sexo.plot(kind='bar', ax = ax)
plt.xlabel('Tipo de régimen de aseguramiento')
plt.ylabel('Frecuencia')
plt.xticks(rotation=0)
plt.xticks(fontsize=9)
plt.title("Caracterizacion de tipo de régimen de aseguramiento fallecidos por sexo",fontsize=15)
#Bucle para que me muestre la etiqueta de cada columna
for container in ax.containers:
    ax.bar_label(container);
```

```python
#Tabla cruzada para causa de muerte por genero
df_genero_causas=pd.crosstab(df['Descripción causa básica de muerte'], df['Sexo'])
df_genero_causas
```

```python
#Grafico de barras distribucion de causas de muerte-genero
fig, ax = plt.subplots(figsize = (14, 4))
ax = df_genero_causas.plot(kind='bar', ax = ax)
plt.xlabel('Causas de muerte')
plt.ylabel('Frecuencia')
plt.xticks(rotation=0)
plt.xticks(fontsize=9)
plt.title("Caracterizacion de causa de muerte entre mujeres y hombres",fontsize=15)
#Bucle para que me muestre la etiqueta de cada columna
for container in ax.containers:
    ax.bar_label(container);
```

```python
#Tabla cruzada para causa de muerte por grupos de edad
df_causas_grupos=pd.crosstab(df['Descripción causa básica de muerte'], df['Grupo de edad'])
df_causas_grupos
```

```python
#Grafico de barras distribucion de pisos
fig, ax = plt.subplots(figsize = (20, 4))
ax = df_causas_grupos.plot(kind='bar', ax = ax)
plt.xlabel('Causas de muerte')
plt.ylabel('Frecuencia')
plt.xticks(rotation=0)
plt.xticks(fontsize=9)
plt.title("Caracterización de causa de muerte entre por grupos de edadaes",fontsize=15)
for container in ax.containers:
    ax.bar_label(container);
```

```python
#Tabla cruzada por genero y año del desceso
df_año_muerte_sexo= pd.crosstab(df['Año'], df['Sexo'])
df_año_muerte_sexo
```

```python
#Grafico de dispersión cantidad muertes por año - genero
#Grafico de barras distribucion de causas de muerte-genero
fig, ax = plt.subplots(figsize = (14, 4))
ax = df_año_muerte_sexo.plot(kind='bar', ax = ax)
plt.xlabel('Periodo(Año)')
plt.ylabel('Frecuencia')
plt.xticks(rotation=0)
plt.xticks(fontsize=9)
plt.title("Caracterizacion por periodos de fallecimiento entre mujeres y hombres",fontsize=15)
#Bucle para que me muestre la etiqueta de cada columna
for container in ax.containers:
    ax.bar_label(container);
```
